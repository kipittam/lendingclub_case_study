#### Outlier Handling

```python
# Function to identify outliers using IQR
def identify_outliers(df):
    outliers = pd.DataFrame()
    for column in df.select_dtypes(include=["float64", "int64"]).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = df[column].apply(
            lambda x: x < lower_bound or x > upper_bound
        )
    return outliers
```

```python
# Identify outliers
outliers = identify_outliers(loan_df)
check_outliers = loan_df[outliers]
check_outliers.dropna(axis=1, how="all", inplace=True)

# Note down outliers in the columns
print(check_outliers.columns)
print(
    "Check outliers annual income:\n",
    check_outliers.annual_inc.value_counts().index.sort_values(ascending=False),
)
print(
    "Check outliers loan amount:\n",
    check_outliers.loan_amnt.value_counts().index.sort_values(ascending=False),
)
print(
    "Check outliers revolving balance:\n",
    check_outliers.revol_bal.value_counts().index.sort_values(ascending=False),
)
print(
    "Check outliers Interest rate:\n",
    check_outliers.int_rate.value_counts().index.sort_values(ascending=False),
)
```

```python
# Function to Filter outliers for a given column


def remove_outliers(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    # print(outliers[column].value_counts())

    # Remove outliers
    df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    return df
```
